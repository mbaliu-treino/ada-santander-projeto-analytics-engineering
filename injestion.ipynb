```python
import pandas as pd

# EXTRA√á√ÉO DE ARQUIVOS LOCAIS (USAR O BANCO DE DADOS)

# reviews = pd.read_csv('./data_aquisition/beta/reviews.csv.gz')
# calendar = pd.read_csv('data_aquisition/beta/calendar.csv.gz')
listings = pd.read_csv('data_aquisition/beta/listings.csv.gz')
listings.shape
```

```python
columns_types = listings.dtypes
dtypes_list = list(columns_types.unique())
columns_types = columns_types.sort_values().to_frame().rename(columns={0: 'dtype'}).reset_index().groupby('dtype')

for i in dtypes_list:
    display(columns_types.get_group(i))
```

```python

def get_duplicated(df: pd.DataFrame, column: str) -> pd.DataFrame:
    return df[df[column].duplicated()]

def monitor_duplicated(df: pd.DataFrame, column: str):
    duplicated_rows = get_duplicated(df, column)

    if len(duplicated_rows) == 0:
        return None
    else:
        return duplicated_rows


cleaning_monitor = {}

column_name = 'id'
sanitizing_duplicated = monitor_duplicated(listings, column_name)
cleaning_monitor[column_name] = {}
if sanitizing_duplicated == None:
    cleaning_monitor[column_name]['duplicated'] = True
else:
    cleaning_monitor[column_name]['duplicated'] = sanitizing_duplicated





# TESTES
# get_duplicated(listings, 'id')
monitor_duplicated(listings, 'id')

```
